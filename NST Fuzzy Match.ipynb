1. filter for just Borno State
2. Filter for locations where community not null
3. run match against each row
4. Append the first score, matched name, ward code, Point x, Point y

```python
df_borno = df_nst.rename(columns={'Community (city,town, ward)': 'ward'})
```

```python
df_borno = df_borno[df_borno['State']=='Borno']
```

```python
df_borno = df_borno[df_borno['ward'].notnull()]
```

```python
df_borno
```

```python
def check_name(ward_name, ward_name_list):
    match = process.extract(ward_name, ward_name_list, limit=1)
    return match[0][0], match[0][1]
```

```python
df_borno['ward_match'], df_borno['match_score'] = df_borno['ward'].apply(lambda x: check_name(x, names))
```
